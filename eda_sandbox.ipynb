```python
import pandas as pd
import numpy as np

# Load the localized Kenyan dataset
df = pd.read_csv('hrm_mock_data.csv')
df.head()
```

```python
# 1. Total Headcount & Baseline Turnover Rate
total_employees = len(df)
attrition_count = df[df['Attrition'] == 'Yes'].shape[0]
attrition_rate = (attrition_count / total_employees) * 100

print("==============================================")
print("          COMPANY-WIDE KEY METRICS            ")
print("==============================================")
print(f"📊 Total Workforce Headcount : {total_employees} employees")
print(f"🚨 Left / Active Flight Risk : {attrition_count} employees")
print(f"📉 Overall Turnover Baseline : {attrition_rate:.1f}%")
print("==============================================")
```

```python
# 2. Regional Distribution Summary
regional_summary = df.groupby('Location').agg(
    Employee_Count=('EmployeeID', 'count'),
    Average_Salary_KES=('Salary', 'mean'),
    Average_Tenure_Years=('TenureYears', 'mean')
).round(1)

print("🏢 REGIONAL OPERATIONAL METRICS")
regional_summary.sort_values(by='Employee_Count', ascending=False)
```

```python
# 3. Turnover Breakdown by Department
dept_turnover = df.groupby('Department')['Attrition'].value_counts(normalize=True).unstack().fillna(0)
if 'Yes' in dept_turnover.columns:
    dept_turnover['Turnover_Rate_%'] = (dept_turnover['Yes'] * 100).round(1)
    print("⚠️ TALENT LEAK BY DEPARTMENT (TURNOVER RATE)")
    display(dept_turnover[['Turnover_Rate_%']].sort_values(by='Turnover_Rate_%', ascending=False))
else:
    print("No attrition detected in the dataset.")
```
